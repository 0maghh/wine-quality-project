```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA

# Charger le dataset
df = pd.read_csv("winequality-red.csv", sep=";")

# Vérifier que ça marche
print("✅ Dataset chargé avec succès !")
print(f"   Nombre de vins     : {df.shape[0]}")
print(f"   Nombre de variables: {df.shape[1]}")
print(f"\nColonnes disponibles :\n{list(df.columns)}")
print(f"\nQualités possibles : {sorted(df['quality'].unique())}")
print(f"\nAperçu des données :")
df.head()
```

```python
print(df.describe().round(2))
```

```python
df['quality'].value_counts().sort_index().plot(kind='bar')
plt.show()
```

```python
df.hist(figsize=(14, 10), bins=20)
plt.show()
```

```python
corr = df.corr()
sns.heatmap(corr, annot=True)
plt.show()
```

```python
variables = ['alcohol', 'volatile acidity', 'sulphates', 'citric acid']

for var in variables:
    df.boxplot(column=var, by='quality')
    plt.show()
```
