```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error

df = pd.read_csv("df_prepared.csv")

# === RÉGRESSION LINÉAIRE SIMPLE ===
print("=== Régression linéaire simple : alcohol → quality ===\n")

X_simple = df[['alcohol']]
y = df['quality']

model_simple = LinearRegression()
model_simple.fit(X_simple, y)

y_pred_simple = model_simple.predict(X_simple)
r2_simple = r2_score(y, y_pred_simple)
rmse_simple = np.sqrt(mean_squared_error(y, y_pred_simple))

print(f"  Coefficient a (pente)  : {model_simple.coef_[0]:.4f}")
print(f"  Intercept b            : {model_simple.intercept_:.4f}")
print(f"  R²                     : {r2_simple:.4f}")
print(f"  RMSE                   : {rmse_simple:.4f}")
print(f"\n  Équation : quality = {model_simple.coef_[0]:.3f} × alcohol + {model_simple.intercept_:.3f}")

```

```python
# Visualisation
plt.figure(figsize=(9, 5))
plt.scatter(df['alcohol'], y, alpha=0.3, s=15, color='steelblue', label='Données')
x_line = np.linspace(df['alcohol'].min(), df['alcohol'].max(), 100).reshape(-1, 1)
plt.plot(x_line, model_simple.predict(x_line), color='red',
         linewidth=2, label=f'Droite de régression (R²={r2_simple:.3f})')
plt.xlabel("Alcool (%)")
plt.ylabel("Qualité")
plt.title("Régression linéaire simple : Alcool → Qualité")
plt.legend()
plt.tight_layout()
plt.savefig("regression_simple.png", dpi=150)
plt.show()
```

```python
# === RÉGRESSION LINÉAIRE MULTIPLE ===
print("\n\n=== Régression linéaire multiple ===\n")

features = ['alcohol', 'volatile acidity', 'sulphates', 'citric acid',
            'total sulfur dioxide', 'density', 'pH']
X_mult = df[features]

X_train, X_test, y_train, y_test = train_test_split(
    X_mult, y, test_size=0.2, random_state=42)

model_mult = LinearRegression()
model_mult.fit(X_train, y_train)

y_pred_train = model_mult.predict(X_train)
y_pred_test  = model_mult.predict(X_test)

print(f"  R² (train)  : {r2_score(y_train, y_pred_train):.4f}")
print(f"  R² (test)   : {r2_score(y_test, y_pred_test):.4f}")
print(f"  RMSE (test) : {np.sqrt(mean_squared_error(y_test, y_pred_test)):.4f}")

print("\n  Coefficients :")
for feat, coef in zip(features, model_mult.coef_):
    print(f"    {feat:30s}: {coef:+.4f}")
```

```python
# Valeurs observées vs prédites
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred_test, alpha=0.4, s=20, color='steelblue')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()],
         'r--', linewidth=1.5, label='Prédiction parfaite')
plt.xlabel("Valeurs observées")
plt.ylabel("Valeurs prédites")
plt.title(f"Régression multiple — R² test = {r2_score(y_test, y_pred_test):.3f}")
plt.legend()
plt.tight_layout()
plt.savefig("regression_multiple_pred.png", dpi=150)
plt.show()

# Importance des variables
coefs = pd.Series(model_mult.coef_, index=features).sort_values()
plt.figure(figsize=(8, 5))
coefs.plot(kind='barh', color=['#E24B4A' if c < 0 else '#1D9E75' for c in coefs])
plt.axvline(0, color='black', linewidth=0.8)
plt.title("Coefficients de la régression multiple")
plt.tight_layout()
plt.savefig("regression_multiple_coefs.png", dpi=150)
plt.show()
```
